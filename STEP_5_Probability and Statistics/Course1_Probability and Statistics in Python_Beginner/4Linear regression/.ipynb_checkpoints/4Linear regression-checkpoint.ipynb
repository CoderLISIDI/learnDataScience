## 2. Drawing lines

```python
# Instructions
# 在我们开始线性回归之前，让我们看看如何绘制线条。
# 一条简单的线 y=x.这意味着y轴上的一个点的值与x轴上的相应值相同。
# 绘制方程 y=x-1,使用现有的x变量。
# 绘制方程 y=x+10,使用现有的x变量。
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4, 5]

# Going by our formula, every y value at a position is the same as the x-value in the same position.
# We could write y = x, but let's write them all out to make this more clear.
y = [0, 1, 2, 3, 4, 5]

# As you can see, this is a straight line that passes through the points (0,0), (1,1), (2,2), and so on.
plt.plot(x, y)
plt.show()

# Let's try a slightly more ambitious line.
# What if we did y = x + 1?
# We'll make x an array now, so we can add 1 to every element more easily.
x = np.asarray([0, 1, 2, 3, 4, 5])
y = x + 1

# y is the same as x, but every element has 1 added to it.
print(y)

# This plot passes through (0,1), (1,2), and so on.
# It's the same line as before, but shifted up 1 on the y-axis.
plt.plot(x, y)
plt.show()

# By adding 1 to the line, we moved what's called the y-intercept -- where the line intersects with the y-axis.
# Moving the intercept can shift the whole line up (or down when we subtract).
y = x - 1
plt.plot(x, y)
plt.show()
y = x + 10
plt.plot(x, y)
plt.show()
```

## 3. Working with slope

```python
# Instructions
# 现在我们有了一种上下移动线路的方法，线路的陡峭程度如何？

# 这一点在之前没有变化 - 每次在x轴上上升1时，y轴上的值总是上升1。

# 如果我们想要在y轴上每行增加2个数字，就会在x轴上增加1个数字？

# 这是斜坡出现的地方。斜率乘以x值得到新的y值。
# 它看起来像这样：y = mx。 如果我们设定斜率m，等于2，我们将得到我们想要的。

# 使用现有的x变量绘制方程y = 4x。

# 使用现有的x变量绘制方程y = .5x。

# 使用现有的x变量绘制方程y = -2x。
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.asarray([0, 1, 2, 3, 4, 5])

# Let's set the slope of the line to 2.
y = 2 * x

# See how this line is "steeper" than before?  The larger the slope is, the steeper the line becomes.
# On the flipside, fractional slopes will create a "shallower" line.
# Negative slopes will create a line where y values decrease as x values increase.
plt.plot(x, y)
plt.show()

y = 4 * x
plt.plot(x, y)
plt.show()

y = .5 * x
plt.plot(x, y)
plt.show()

y = -2 * x
plt.plot(x, y)
plt.show()
```

## 4. Starting out with linear regression

```python
# Instructions
# 在上一次任务中，我们用r值做了一些工作。 r值表示两个变量是如何相关的。 这可以从不相关到负相关到正相关。

# 两个变量越相关，使用一个来预测另一个变得越容易。 
# 例如，如果我知道我为我的牛排付了多少与牛排的大小（盎司）高度正相关，我可以创建一个公式，帮助我预测我将为我的牛排付多少钱。

# 我们这样做的方式是线性回归。 线性回归给我们一个公式。 如果我们把一个变量的值插入到这个公式中，我们得到另一个变量的值。

# 创建公式的公式采用y = mx + b的形式。

# 您可能会从过去的两个屏幕中识别出这个方程的一部分 - 我们只是将截距和斜率添加到一个方程中。

# 这个公式表示“第二个变量（y）的预测值等于第一个变量（x）乘以斜率（m）加上截距（b）”的值。

# 在使用我们的公式之前，我们必须计算m和b的值。

# 我们首先计算斜率 - 公式是cov（x，y）/ σx2，它只是x和y的协方差除以x的方差。

# 我们可以使用cov函数来计算协方差，并且使用Pandas系列上的.var（）方法来计算方差。

#     计算使用“密度”列（x）预测“质量”列（y）所需的斜率。

#     将斜率分配给slope_density。
```

```python
# The wine quality data is loaded into wine_quality

from numpy import cov
import pandas as pd
wine_quality = pd.read_csv('wine_quality_white.csv')
slope_density = cov(wine_quality["density"], wine_quality["quality"])[0, 1] / wine_quality["density"].var()
print(slope_density)
```

## 5. Finishing linear regression

```python
# Instructions
# 现在我们可以计算我们的线性回归线的斜率，我们只需要计算截距。
# 截距就是平均y点比我们的预测值高或低多少。

# 我们可以通过我们计算过的斜率来计算截距：y¯-mx¯。 所以我们就取y值的均值，然后从中减去斜率乘以x值的平均值。
# 请记住，我们可以通过使用.mean（）方法来计算均值。

# 使用“密度”列（x）计算您需要预测“质量”列（y）的y截距。
# 将结果分配给intercept_density。
```

```python
from numpy import cov
# This function will take in two columns of data, and return the slope of the linear regression line.

def calc_slope(x, y):
    return cov(x, y)[0, 1] / x.var()

intercept_density = wine_quality["quality"].mean() - (calc_slope(wine_quality["density"], wine_quality["quality"]) 
                                                      * wine_quality["density"].mean())
print(intercept_density)
```

## 6. Making predictions

```python
# Instructions
# 现在我们已经计算了斜率和截距，我们可以从x值中预测y值。

# 为了做到这一点，我们回到原来的公式：y = mx + b，并插入m和b的值。

# 然后，我们可以计算任何x值的预测y值。 这让我们可以预测我们从未见过的x值的质量。 
# 例如，密度为.98的葡萄酒不在我们的数据集中，但是我们可以预测评论者将使用该密度分配给葡萄酒的质量。

# 根据预测变量和被预测值的关联程度，预测可能是好的或坏的。

# 现在让我们看看做出预测，然后我们将继续研究它们有多好。

#     编写一个函数来计算给定x值的预测y值。

#     使用“密度”列上的.apply（）方法将函数应用于列中的每个项目。 这将计算所有预测的y值。

#     将结果分配给prediction_quality。
```

```python
from numpy import cov

def calc_slope(x, y):
    return cov(x, y)[0, 1] / x.var()

# Calculate the intercept given the x column, y column, and the slope
def calc_intercept(x, y, slope):
    return y.mean() - (slope * x.mean())

slope = calc_slope(wine_quality["density"], wine_quality["quality"])
intercept = calc_intercept(wine_quality["density"], wine_quality["quality"], slope)

def compute_predicted_y(x):
    return x * slope + intercept

predicted_quality = wine_quality["density"].apply(compute_predicted_y)
print(predicted_quality)
```

## 7. Finding error

```python
# Instructions
# 现在我们知道如何手动创建一个回归线，让我们来看一个更简单的方法来使用scipy中的函数。

# linregress函数使线性回归变得简单。

# 现在我们知道了一种更简单的线性回归方法，让我们看看如何确定我们的回归是好还是坏。

# 我们可以画出我们的线条和我们的实际值，看看它们在y轴上有多远。

# 我们还可以计算每个预测与实际值之间的距离 - 这些距离称为残差。

# 如果我们加上残差的平方和，我们可以得到一个很好的误差估计。

# 我们必须加上平方残差，因为就像平均值的差异一样，残差如果不是平方的，就加到0。
# 用数学术语来说，残差的平方和为：
```

```python
from IPython.display import Image
Image(filename='1.png')
```

```python
# 使用给定的斜率和截距，计算预测的y值。

# 从相应的实际y值中减去每个预测的y值，平方差，并将所有的差值加在一起。

# 这会给你残差的平方和。 将此值分配给rss。
```

```python
from scipy.stats import linregress

# We've seen the r_value before -- we'll get to what p_value and stderr_slope are soon -- for now, don't worry about them.
slope, intercept, r_value, p_value, stderr_slope = linregress(wine_quality["density"], wine_quality["quality"])

# As you can see, these are the same values we calculated (except for slight rounding differences)
print(slope)
print(intercept)

import numpy
predicted_y = numpy.asarray([slope * x + intercept for x in wine_quality["density"]])
residuals = (wine_quality["quality"] - predicted_y) ** 2
rss = sum(residuals)
```
