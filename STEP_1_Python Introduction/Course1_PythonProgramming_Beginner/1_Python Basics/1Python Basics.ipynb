```python
# Learn--
# 现在我们知道如何创建一个列表并为其添加值，我们来学习如何访问和使用我们创建的列表中的值。 
# 列表中的每个值都有一个与之关联的索引或位置。 
# 列表从索引0开始，一直小于列表中值或元素的总数。 
# 例如，如果您有一个包含5个值的列表，则索引的范围将从0到4。
# 列表索引的主要缺点是要访问列表中的第一个元素，我们实际上使用的索引值为0，不是1.
# 第二个元素的索引值为1，第三个元素的索引值为2，依此类推。 
# 这被称为零索引。 
# 虽然许多编程语言都使用零索引，但是像MATLAB这样的编程语言却没有。
```

```python
from IPython.display import Image
Image(filename='python_list_indexing.png')
```

```python
# 要返回具有给定索引的值，请将索引的整数传递给括号表示法。 
# 在下面的代码中，我们访问列表crime_rates中的第一个，第二个和第五个元素。 
# 我们将每个访问的值分配给新的变量：
```

```python
crime_rates = [749, 371, 828, 503, 1379]
first_value = crime_rates[0]
second_value = crime_rates[1]
fifth_value = crime_rates[4]
```
