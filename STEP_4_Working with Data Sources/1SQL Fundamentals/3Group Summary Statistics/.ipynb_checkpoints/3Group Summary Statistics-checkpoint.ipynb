## 3. Practice: Using GROUP BY

```python
# 现在我们对GROUP BY语句有了更好的理解，让我们通过为recent_grads表按组计算摘要统计信息来练习使用它。
```

```python
# 对于每个主要类别，查找受雇毕业生的百分比。
# 使用SELECT语句在查询中选择以下列和聚合：
#     Major_category
#     AVG（就业）/ AVG（合计）作为share_employed
# 使用GROUP BY语句按Major_category列对查询进行分组。
```

```python
# SELECT Major_category, AVG(Employed) / AVG(Total) AS share_employed FROM recent_grads GROUP BY Major_category;
```

## 4. Querying Virtual Columns With the HAVING Statement

```python
# 有时我们想要在执行GROUP BY查询后选择行的子集。 例如，在最后一个屏幕上，我们可能只想选择share_employed大于.8的那些行。 
# 我们不能使用WHERE子句来执行此操作，因为share_employed不是recent_grads中的列; 
# 它实际上是一个由GROUP BY语句生成的虚拟列。
# 当我们想要过滤由查询生成的列时，我们可以使用HAVING语句。 这是一个例子：
```

```python
# SELECT Major_category, AVG(Employed) / AVG(Total) AS share_employed 
# FROM recent_grads 
# GROUP BY Major_category 
# HAVING share_employed > .8;
```

## 5. Rounding Results With the ROUND() Function

```python
# 在最后一个屏幕上，我们的结果中的百分比非常长，很难理解（例如，0.16833085991095678）。 
# 我们可以在查询中使用SQL ROUND函数对它们进行舍入。 
# 下面是一个这样的例子：
```

```python
# SELECT ROUND(ShareWomen, 4), Major_category FROM recent_grads LIMIT 10;
```

```python
# SELECT Major_category, ROUND(AVG(College_jobs) / AVG(Total), 3) AS share_degree_jobs 
# FROM recent_grads 
# GROUP BY Major_category 
# HAVING share_degree_jobs < .3;
```

## 6. Nesting functions

```python
SELECT Major_category, ROUND(AVG(College_jobs) / AVG(Total), 3) AS share_degree_jobs 
FROM recent_grads 
GROUP BY Major_category 
HAVING share_degree_jobs < .3;
```
